```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
with open('pricing test data (3).csv', 'r', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    l = []
    k = 0
    for row in reader:
        if k == 0:
            pass
        else:
            l.append([float(row[0], )] + row[1:])
        k+=1
    df = pd.DataFrame(l)
df.columns = ['user id','event_time','event name','group','country','source']
df.to_csv('1.csv', index=False)
df
```
